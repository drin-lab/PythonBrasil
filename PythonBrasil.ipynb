```python
paises_python = country_grupo['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
python_df.loc['Brazil']
```

```python
python_df.rename(columns={'Country':'NumRespondentes','LanguageHaveWorkedWith':'UsuariosPython'}, inplace=True)
```

```python
python_df['%UsuariosPython'] = (python_df['UsuariosPython']/python_df['NumRespondentes'])*100
```
